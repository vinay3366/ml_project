```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC 
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from xgboost import XGBClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, roc_auc_score, classification_report, roc_curve
from sklearn.metrics import confusion_matrix
```

```python
dataset=pd.read_csv('ASD.csv')
```

```python
print(dataset.dtypes)
```

```python
print(dataset.info())
```

```python
print(dataset.describe())
```

```python
dataset.replace("?", np.nan, inplace= True)
print(dataset.isnull().sum())
```

```python
dataset['Age']=dataset['Age'].fillna(dataset['Age'].mode()[0])
dataset['Ethincity']=dataset['Ethincity'].fillna(dataset['Ethincity'].mode()[0])
dataset['Who completed the test']=dataset['Who completed the test'].fillna(dataset['Who completed the test'].mode()[0])
```

```python
X = dataset.iloc[:,1:17].values
print(X[:10])
```

```python
y = dataset.iloc[:,17].values
print(y[:10])
```

```python
label_encoder=LabelEncoder()
X[ :,11]=label_encoder.fit_transform(X[ :,11])
X[ :,12]=label_encoder.fit_transform(X[ :,12])
X[ :,13]=label_encoder.fit_transform(X[ :,13])
X[ :,14]=label_encoder.fit_transform(X[ :,14])
X[ :,15]=label_encoder.fit_transform(X[ :,15])
y=label_encoder.fit_transform(y)
print(X[:50])
print(y[:70])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30) 
```

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)

y_pred = lr.predict(X_test)
print(accuracy_score(y_test,y_pred))
```

```python
cm= confusion_matrix(y_test, y_pred)
ax= plt.subplot()
sns.heatmap(cm, annot=True, ax = ax); #annot=True to annotate cells
print(confusion_matrix(y_test, y_pred))
# labels, title and ticks
ax.set_xlabel('Predicted labels');ax.set_ylabel('True labels'); 
ax.set_title('Confusion Matrix for Logistic Regression'); 
ax.xaxis.set_ticklabels(['negative', 'positive']); ax.yaxis.set_ticklabels(['negative', 'positive']);
```

```python
svclassifier = SVC(kernel='linear')  
svc=svclassifier.fit(X_train, y_train)  

y_pred = svclassifier.predict(X_test)
print(accuracy_score(y_test,y_pred))
```

```python
nb=GaussianNB()
nb.fit(X_train,y_train)

y_pred = nb.predict(X_test)
print(accuracy_score(y_test,y_pred))
```

```python
cm= confusion_matrix(y_test, y_pred)
ax= plt.subplot()
sns.heatmap(cm, annot=True, ax = ax); #annot=True to annotate cells
print(confusion_matrix(y_test, y_pred))
# labels, title and ticks
ax.set_xlabel('Predicted labels');ax.set_ylabel('True labels'); 
ax.set_title('Confusion Matrix for NAIVE BAYES'); 
ax.xaxis.set_ticklabels(['negative', 'positive']); ax.yaxis.set_ticklabels(['negative', 'positive']);
```

```python
gender1={'m':0,'f':0}
gender2={'m':0,'f':0}
gender3={'m':0,'f':0}
gender4={'m':0,'f':0}
i=0
jaundice={'yes':0,'no':0}
for a  in dataset['Jaundice']:
    b=dataset['Sex'].values[i].lower()
    c=dataset['Class'].values[i].lower()
    i=i+1
    jaundice[a]+=1
    if(a=='yes' and c=='yes'):
        gender1[b]+=1
    elif(a=='yes' and c=='no'):
        gender2[b]+=1
    elif(a=='no' and c=='yes'):
        gender3[b]+=1
    else:
        gender4[b]+=1
print(jaundice['yes'],jaundice['no'])
print('ASD & JAUNDICE    : ',gender1['m'],gender1['f'],'\tTOTAL: ',gender1['m']+gender1['f'])
print('ASD & NO JAUNDICE : ',gender2['m'],gender2['f'],'\tTOTAL: ',gender2['m']+gender2['f'])
print('JAUNDICE & No ASD : ',gender3['m'],gender3['f'],'\tTOTAL: ',gender3['m']+gender3['f'])
print('No ASD & JAUNDICE : ',gender4['m'],gender4['f'],'\tTOTAL: ',gender4['m']+gender4['f'])
```
